## Import libraries

```python
import pandas as pd
import numpy as np
import math
import os
import csv
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import mean_squared_error
import statsmodels.api as sm
from scipy.stats import pearsonr
```

## Read data

```python
df = pd.read_csv('insurance.csv')
df
```

## Data processing

### Check null values in dataset

```python
df.isna().sum()
```

### Check categorical column values

```python
cat_columns = ['sex', 'region', 'smoker']

for i in cat_columns:
    print(df[i].unique())
```

### Dummy encoding 

```python
df['sex'] = pd.get_dummies(df['sex'])
df['smoker'] = pd.get_dummies(df['smoker'])
df
```

```python
# for region, use one-hot encoding

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df['region'])
print(df['region'].head(10))

# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)

print(onehot_encoded.shape)
print(onehot_encoded)
```

```python
# add one-hot encoded variables to dataset

df['northeast'] = onehot_encoded[:, 0]
df['northwest'] = onehot_encoded[:, 1]
df['southeast'] = onehot_encoded[:, 2]
df['southwest'] = onehot_encoded[:, 3]
df
```

### Data analysis 

```python
# Check correlation with target variable
rows = df.loc[:, ~df.columns.isin(['region', 'charges'])].columns
columns = ['correlation']
index = [rows]
corr_df = pd.DataFrame(columns=columns, index=rows)

for count, val in enumerate(rows):
    corr, _ = pearsonr(df[val], df['charges'])
    corr_df.iloc[count, 0] = abs(corr)

corr_df
```

```python
# Move qualitative/ target col to the end for ease of indexing

cols = ['region', 'charges']

for i in cols:
    move_col = df.pop(i)

    df.insert(len(df.columns), i, move_col)
    
df
```

```python
# Check correlation between variables to check for multicollinearity

f,ax = plt.subplots(figsize=(18,18))
plt.title('Correlation Matrix')
sns.heatmap(df.iloc[:,:-2].corr(method='pearson'), annot=True, fmt = '.2f', ax=ax)
plt.show()
```

## Linear Regression Testing

```python
# create function to run single-factor linear regression model for the data

def single_factor(dataframe):
    df_num = dataframe.select_dtypes(include=[np.float, np.int64, np.uint8])
    index = df_num.columns[:-1]
    cols = ['slope', 'intercept', 'r2', 'p_val']
    linear_reg_results = pd.DataFrame(columns=cols, index=index)
    
    for count, val in enumerate(index):
        x = df_num.iloc[:, count]
        y = df_num.iloc[:, -1]
        x_ = np.array(x).reshape(-1, 1)
        y_ = np.array(y)
        model = LinearRegression().fit(x_, y_)
        slope = model.coef_
        intercept = model.intercept_
        r_sq = model.score(x_, y_)
        
        x = sm.add_constant(x)
        mod = sm.OLS(y,x)
        mod_fit = mod.fit()
        p_value = round(mod_fit.pvalues[1], 2)
        
        linear_reg_results.iloc[count, 0] = slope
        linear_reg_results.iloc[count, 1] = intercept
        linear_reg_results.iloc[count, 2] = r_sq
        linear_reg_results.iloc[count, 3] = p_value
        
        linear_reg_results.sort_values('r2', ascending=False, inplace=True)

    return linear_reg_results

single_factor_regressions = single_factor(df)
single_factor_regressions
```

```python
# create function to run two-factor linear regression model for the data


def two_factor(dataframe):
    df_num = dataframe.select_dtypes(include=[np.float, np.int64, np.uint8])
    cols = ['x1', 'x2', 'slope_1', 'slope_2', 'intercept', 'r2', 'p1', 'p2']
    linear_reg_results = pd.DataFrame(columns=cols)
    iter_cols = df.columns[:-2]
    
    for count, i in enumerate(iter_cols):
        df_new = df_num.drop(i, axis=1).iloc[:, :-1]
        for j in df_new.columns:
            x = df_num[[i, j]]
            y = df_num.iloc[:, -1]
            
            model = LinearRegression().fit(x, y)
            slope_1 = model.coef_[0]
            slope_2 = model.coef_[1]
            intercept = model.intercept_
            r_sq = model.score(x, y)
            
            x = sm.add_constant(x)
            mod = sm.OLS(y,x)
            mod_fit = mod.fit()
            p_value_1 = round(mod_fit.pvalues[1], 2)
            p_value_2 = round(mod_fit.pvalues[2], 2)
            
            new_df = pd.DataFrame({'x1': i, 'x2': j, 'slope_1': slope_1, 'slope_2': slope_2,
                                                'intercept': intercept, 'r2': r_sq, 'p1': p_value_1, 'p2': p_value_2}, index=[0])
            
            linear_reg_results = pd.concat([linear_reg_results, new_df], axis=0)
    
    linear_reg_results.drop_duplicates(subset=["r2"], inplace=True)

    linear_reg_results.sort_values("r2", ascending=False, inplace=True)  
            
    return linear_reg_results


two_factor_reg = two_factor(df)
two_factor_reg
```

```python
# create function to run three-factor linear regression model for the data

def three_factor(dataframe):
    df_num = dataframe.select_dtypes(include=[np.float, np.int64, np.uint8])
    cols = ['x1', 'x2', 'x3', 'slope_1', 'slope_2', 'slope_3', 'intercept', 'r2', 'p1', 'p2', 'p3']
    linear_reg_results = pd.DataFrame(columns=cols)
    iter_cols = df.columns[:-2]
    
    for count, i in enumerate(iter_cols):
        df_new = df_num.drop(i, axis=1).iloc[:, :-1]
        for j in df_new.columns:
            df_new2 = df_new.drop(j, axis=1).iloc[:, :-1]
            for k in df_new2.columns:
                x = df_num[[i, j, k]]
                y = df_num.iloc[:, -1]

                model = LinearRegression().fit(x, y)
                slope_1 = model.coef_[0]
                slope_2 = model.coef_[1]
                slope_3 = model.coef_[2]
                intercept = model.intercept_
                r_sq = model.score(x, y)

                x = sm.add_constant(x)
                mod = sm.OLS(y,x)
                mod_fit = mod.fit()
                p_value_1 = round(mod_fit.pvalues[1], 2)
                p_value_2 = round(mod_fit.pvalues[2], 2)
                p_value_3 = round(mod_fit.pvalues[3], 2)

                new_df = pd.DataFrame({'x1': i, 'x2': j, 'x3': k, 'slope_1': slope_1, 'slope_2': slope_2, 'slope_3': slope_3, 
                                       'intercept': intercept, 'r2': r_sq, 'p1': p_value_1, 'p2': p_value_2, 'p3': p_value_3}, index=[0])

                linear_reg_results = pd.concat([linear_reg_results, new_df], axis=0)
    
    linear_reg_results.drop_duplicates(subset=["r2"], inplace=True)

    linear_reg_results.sort_values("r2", ascending=False, inplace=True)  
            
    return linear_reg_results


three_factor_reg = three_factor(df)
three_factor_reg
```

```python
datasets = [three_factor_reg, two_factor_reg, single_factor_regressions]

# Print the best model for each regression type, to check the overall best linear regression model (based on the R-squared)

for i in datasets:
    print(i.iloc[0, :])
```

### RMSE Calculation

```python
# Check mean squared error for each model

def root_mean_squared_error(dataframe, *variables):
    if len(variables) == 3:
        x = dataframe[[variables[0], variables[1], variables[2]]]
    elif len(variables) == 2:
        x = dataframe[[variables[0], variables[1]]]
    else:
        x = dataframe[variables[0]]
        x = np.array(x).reshape(-1, 1)
        
    y = dataframe['charges']
    
    model = LinearRegression().fit(x, y)
    
    y_pred = model.predict(x)
    
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    
    return rmse

rmse_df = pd.DataFrame(columns=['x1', 'x2', 'x3', 'RMSE'], index=['three-factor', 'two-factor', 'single-factor'])

for i in datasets:
    if i is three_factor_reg:
        x1 = three_factor_reg.iloc[0, 0]
        x2 = three_factor_reg.iloc[0, 1]
        x3 = three_factor_reg.iloc[0, 2]
        rmse_df.iloc[0, 0] = x1
        rmse_df.iloc[0, 1] = x2
        rmse_df.iloc[0, 2] = x3
        rmse_df.iloc[0, 3] = root_mean_squared_error(df, x1, x2, x3)
    elif i is two_factor_reg:
        x1 = two_factor_reg.iloc[0, 0]
        x2 = two_factor_reg.iloc[0, 1]
        rmse_df.iloc[1, 0] = x1
        rmse_df.iloc[1, 1] = x2
        rmse_df.iloc[1, 3] = root_mean_squared_error(df, x1, x2)
    else:
        x1 = single_factor_regressions.index[0]
        rmse_df.iloc[2, 0] = x1
        rmse_df.iloc[2, 3] = root_mean_squared_error(df, x1)
        
rmse_df
```

## Summary
